```python
# tabel = [{"cod produs" : ""      , "denumire vin" : ""      , "Categorie" : ""      , "producător" : ""      , "țara" : ""      , "zonă" : ""      , "apelațiune" : ""      , "soi" : ""      , "an" : ""      , "culoare" : ""      , "tip vin" : ""      , "zahăr rezidual" : ""      , "alcool" : ""      , "capacitate ambalaj" : ""      , "maturare lemn" : ""      , "asociere mâncare" : ""      , "potențial învechire" : ""      , "caracter" : ""      , "temperatură de servire" : ""      , "bio" : ""      , "aerare" : ""      , "RP" : ""      , "JS" : ""      , "JR" : ""      , "AG" : ""      , "D" : ""      , "WS" : ""      , "LM" : ""      , "RVF" : ""      , "90plus" : ""      , "preț" : ""      , "promotie" : ""      , "null"     : ""      , "stoc" : ""      , "poza" : ""      , "descriere" : ""      , "Descriere crama": ""  }]
tabel = []
# for link in links:
    # URL = link
    # resp = requests.get(URL)
for resp in resps:
    # wine = wineHeader
    wine = {"cod produs" : ""      , "denumire vin" : ""      , "Categorie" : ""      , "producător" : ""      , "țara" : ""      , "zonă" : ""      , "apelațiune" : ""      , "soi" : ""      , "an" : ""      , "culoare" : ""      , "tip vin" : ""      , "zahăr rezidual" : ""      , "alcool" : ""      , "capacitate ambalaj" : ""      , "maturare lemn" : ""      , "asociere mâncare" : ""      , "potențial învechire" : ""      , "caracter" : ""      , "temperatură de servire" : ""      , "bio" : ""      , "aerare" : ""      , "RP" : ""      , "JS" : ""      , "JR" : ""      , "AG" : ""      , "D" : ""      , "WS" : ""      , "LM" : ""      , "RVF" : ""      , "90plus" : ""      , "preț" : ""      , "promotie" : ""      , "null"     : ""      , "stoc" : ""      , "poza" : ""      , "descriere" : ""      , "Descriere crama": ""  }

    if resp != "not found" and resp.status_code == 200 :
        # if"search&search" not in link:
        # if resp != "not found":
        soup = BeautifulSoup(resp.text, "lxml")
        data = soup.find(id="productDescription").get_text()
        data = data.replace("\xa0", " ")
        data = data.strip() 
        data = data.split("\n")
        new_data = []
        for line in data:
            line = line.split(":")
            new_data.append(line)
            if line[0] == "Producător":
                wine['producător'] = line[1]
            # elif line[0] == "Perioada optimă de consum":
            #     wine['potențial învechire'] = line[1]
            elif line[0] == "Soi":
                # wine['soi'] = line[1]
                wine['soi'] = re.sub(r'-.*',"",re.sub(r'[0-9]', '', line[1].replace(" ","").replace(",","|").replace("%","").replace("","")))
            elif line[0] == "Alcool":
                wine['alcool'] = line[1]
            elif line[0] == "Temperatura de servire":
                wine['temperatură de servire'] = line[1]
            # if line[0] == "":
            #   wine['Descriere crama'] = line[1]
            # if line[0] == "Recomandări gastronomice":
            #     wine['asociere mâncare'] = line[1]

        wine['asociere mâncare'] = "brânzeturi maturate, carne roșie, vânat"
        wine['capacitate ambalaj'] = "0.75 l"
        wine['potențial învechire'] = " >10 ani"
        
        #  numele cramei / producatorului
        wine['producător'] = soup.find("div", {"class": "manuf_title"}).get_text()
        soup.find('div', {"class": "manuf_title"}).decompose()
        # numele vinului
        denumireVin = soup.find("h1", {"class": "productsTitle"}).get_text().strip()
        wine['denumire vin'] = wine['producător'] + " " + denumireVin[-4:]
        
        # stergem headingul care zice "Producator"
        soup.find('p', {"class": "blueHeadingInner"}).decompose()
        #  luam descrierea cramei / producatorului
        if soup.find(id="productManufacturer"):
            wine['Descriere crama'] = soup.find(id="productManufacturer").get_text().replace("\n"," ").replace("\t","").replace("\r"," ")


        # wine['descriere'] = soup.find('div', {"class": "productInfoModelWrapper"}).get_text().replace("\n"," ").replace("\t","").replace("\r"," ")
        wine['descriere'] = soup.find('div', {"id": "productDescription"})
        # wine['cod produs'] = wine[]
        imageTag = re.findall('src=".*"',str(soup.find('div', {"id": "productMainImage"})))[0][5:-1]
        wine['cod produs'] = unidecode.unidecode(denumireVin.replace(" ","-"))
        wine['poza'] = wine['cod produs']  + imageTag[-4:]
        wine['țara'] = "Franța"
        wine['zonă'] = "Bordeaux"


        # print(new_data)
        # print(wine)
        tabel.append(wine)
        # print(new_data.index("Perioada optimă de consum"))
        print(wine['denumire vin'])
    else:
        wine['cod produs'] = "not found"

        tabel.append(wine)
        print("not found")
        # print(wine)

```
